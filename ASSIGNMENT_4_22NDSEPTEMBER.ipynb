## splitting Dependent and independent variables

```python
df.head()
```

```python
#independet variable
x=df.drop("Attrition",axis=1)
x
```

```python
#y=dependent variable
y=df["Attrition"]
y.columns="Attrition"
y
```

## LabelEncoding

```python
df.head()
```

```python
df.BusinessTravel.value_counts()
```

```python
df.Department.value_counts()
```

```python
df.EducationField.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
#we use LabelEncoder to cahne the categorical to numerical
x.BusinessTravel=le.fit_transform(x.BusinessTravel)
x.Department=le.fit_transform(x.Department)
x.EducationField=le.fit_transform(x.EducationField)
x.Gender=le.fit_transform(x.Gender)
x.Over18=le.fit_transform(x.Over18)
x.OverTime=le.fit_transform(x.OverTime)
x.JobRole=le.fit_transform(x.JobRole)
x.MaritalStatus=le.fit_transform(x.MaritalStatus)

x.head()
```

```python
x.info()
```

```python
y.head()
```

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
x_scaled=pd.DataFrame(ms.fit_transform(x),columns=x.columns)
```

```python
x_scaled.head()
```

## splitting into train test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## MODELBUILDING

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
y_pred.shape
```

```python
y_test
```

## evaluation of metrics

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_curve
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
pd.crosstab(y_test,y_pred)
```

```python
probability=lr.predict_proba(x_test)[:,1]
probability
```

# **MODEL BUILDING USING DECISION TREE CLASSIFIER **

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
pred=dtc.predict(x_test)
pred
```

```python
y_test
```

## Evaluation of classification  model

```python
#accuracy score
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
pd.crosstab(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

## Hyper parameter tuning

```python
from sklearn import tree
plt.figure(figsize=(25,15))
tree.plot_tree(dtc,filled=True)
```

```python
from sklearn.model_selection import GridSearchCV
parameter={
 'criterion':['gini','entropy'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']

}
```

```python
grid_search=GridSearchCV(estimator=dtc,param_grid=parameter,cv=5,scoring="accuracy")
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
dtc_cv=DecisionTreeClassifier(criterion= 'entropy',
 max_depth=3,
 max_features='sqrt',
 splitter='best')
dtc_cv.fit(x_train,y_train)
```

```python
pred1=dtc_cv.predict(x_test)
pred1
```

```python
print(classification_report(y_test,pred1))
```
